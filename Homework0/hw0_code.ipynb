```python
#%pylab inline
import numpy
import csv

# N×p array, where N is the number of samples and p is the number of attributes per sample
matrix = numpy.loadtxt(open("iris.data", "rb"), delimiter=",", usecols=range(4))

# N-dimensional vector containing each sample’s label (species)
labels = [None] * 150
with open('iris.data', 'r') as f:
    reader = csv.reader(f)
    temp = list(reader)    

for i in range(150):
    labels[i] = temp[i][4]
    
from matplotlib import pylab as plt
sepLen = matrix[:,0]
sepWid = matrix[:,1]
petLen = matrix[:,2]
petWid = matrix[:,3]
colors = ["r"] * 150
for i in range(50, 100):
    colors[i] = "g"
for i in range(100, 150):
    colors[i] = "b"
    
f, axarr = plt.subplots(4, 4)
f.suptitle('Iris Data (red=setosa, green=versicolor, blue=virginica)', fontsize=30, fontweight='bold')
f.set_size_inches(16,16)
f.tight_layout(pad=0.5, w_pad=0.5, h_pad=1)
axarr[0, 0].annotate('Sepal.Length', xy=(0.5, 0.5), xycoords='axes fraction', fontsize=30, horizontalalignment='center', verticalalignment='center')
axarr[0, 1].scatter(sepWid, sepLen, c=colors)
axarr[0, 2].scatter(petLen, sepLen, c=colors)
axarr[0, 3].scatter(petWid, sepLen, c=colors)
axarr[1, 0].scatter(sepLen, sepWid, c=colors)
axarr[1, 1].annotate('Sepal.Width', xy=(0.5, 0.5), xycoords='axes fraction', fontsize=30, horizontalalignment='center', verticalalignment='center')
axarr[1, 2].scatter(petLen, sepWid, c=colors)
axarr[1, 3].scatter(petWid, sepWid, c=colors)
axarr[2, 0].scatter(sepLen, petLen, c=colors)
axarr[2, 1].scatter(sepWid, petLen, c=colors)
axarr[2, 2].annotate('Petal.Length', xy=(0.5, 0.5), xycoords='axes fraction', fontsize=30, horizontalalignment='center', verticalalignment='center')
axarr[2, 3].scatter(petWid, petLen, c=colors)
axarr[3, 0].scatter(sepLen, petWid, c=colors)
axarr[3, 1].scatter(sepWid, petWid, c=colors)
axarr[3, 2].scatter(petLen, petWid, c=colors)
axarr[3, 3].annotate('Petal.Width', xy=(0.5, 0.5), xycoords='axes fraction', fontsize=30, horizontalalignment='center', verticalalignment='center')
plt.savefig("iris.png")
```
